# Demo for reading OADAT

```python
import sys, os 
if os.path.basename(os.getcwd()) == 'notebooks':
    os.chdir('..') ## set working dir to root
import matplotlib.pyplot as plt

if 'src' not in sys.path:
    sys.path.append('src')

import generators
```

```python
datasets_parent_dir = '/home/firat/docs/dlbirhoui/parsed_data2'
fname_dataset = 'SWFD_semicircle_RawBP.h5'
fname_h5 = os.path.join(datasets_parent_dir, fname_dataset)
inds = None # if not None, generator will be limited to the provided dataset indices.
in_key = 'sc,ss32_BP'
out_key = 'sc_BP'
gen = generators.Generator_Paired_Input_Output(fname_h5=fname_h5, inds=inds, in_key=in_key, out_key=out_key, shuffle=True)
index_to_retrieve = 42
x, y = gen[index_to_retrieve]
```

```python
fig, axs = plt.subplots(1,2, figsize=(8, 4), dpi=300)
axs[0].imshow(x, cmap='gray')
axs[1].imshow(y, cmap='gray')
axs[0].set_title(in_key)
axs[1].set_title(out_key)
axs[0].set_axis_off()
axs[1].set_axis_off()
plt.show()
```

### Alternatively, data loader can also be used as a generator object

```python
for x,y in gen:
    fig, axs = plt.subplots(1,2, figsize=(8, 4), dpi=300)
    axs[0].imshow(x, cmap='gray')
    axs[1].imshow(y, cmap='gray')
    axs[0].set_title(in_key)
    axs[1].set_title(out_key)
    axs[0].set_axis_off()
    axs[1].set_axis_off()
    plt.show()
    break
```
